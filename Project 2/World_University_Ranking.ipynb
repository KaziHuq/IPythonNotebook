```python
import pandas as pd

data = pd.read_csv('world_university_ranking.csv')

print(data)
```

```python
#Getting the mid point value of Research Score
ResearchScore = (data['Research Score']).median() 
print("Mid point value of Research Score = ",ResearchScore)
```

```python
#Getting the avg of Research Score
TeachingScore = (data['Teaching Score']).mean() 
print("Average of Teaching Score = ",TeachingScore)
```
