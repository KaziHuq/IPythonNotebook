```python
import pandas as pd
import numpy as np

data = pd.read_csv('Movie Ratings.csv')
data.rename( columns={'Unnamed: 0':'User'}, inplace=True )

data
```

```python
##create DataFrame
df = pd.DataFrame({'Spider-Man: Across the Spider-Verse': [4.5, 3.5, 4.8, 3.5],
                   'John Wick: Chapter 4': [4.0, 4.2, 3.8, 3.5],
                   'Ant-Man and the Wasp: Quantumania': [4.5, 3.5, 3.5, 4.2],
                    'Barbie': [3.8, 4.5, 3.0, 4.5],
                   'Oppenheimer': [4.8, 3.8, 4.5, 4.8]
                  
                  })
df
```

```python
#define new column that shows the average row value for all columns
df['average_all'] = df.mean(axis=1)

#view updated DataFrame
df
```

```python
#Data normalization

normal = lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x))

normalized = data.copy()

columns_normal = ["Spider-Man: Across the Spider-Verse", "John Wick: Chapter 4", "Ant-Man and the Wasp: Quantumania", "Barbie", "Oppenheimer"]
normalized = normalized[columns_normal].apply(normal)
normalized
```

```python
#Calculation average rating for all the movies after data normalization
avg = normalized.mean(axis=1)
avg
```

```python
#Data standardization

standard = lambda x: (x-x.mean()) / x.std()

standardized = normalized[columns_normal].apply(standard)
standardized
```

```python
#Calculation average rating for all the users after data standardization
avgUser = standardized.mean(axis='columns')
avgUser
```

```python
#Calculation average rating for all the movies after data standardization
avgMovie = standardized.mean(axis='index')
avgMovie
```
