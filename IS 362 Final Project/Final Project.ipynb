```python
import pandas as pd
import matplotlib.pyplot as plt

'/Understanding the data/'

df = pd.read_csv('wages_by_education.csv')
df.head()
```

```python
'/Cleaning the data/'
df.isnull().sum()
```

```python
'/ US Hourly Wage for the past 50 years/'

axes = plt.gca()
plt.title("Men Hourly Wage Based on Education from 1973 to 2022") 
df.plot(kind='line', x='year', y='men_less_than_hs', ax=axes);
df.plot(kind='line', x='year', y='men_high_school', ax=axes);
df.plot(kind='line', x='year', y='men_some_college', ax=axes);
df.plot(kind='line', x='year', y='men_bachelors_degree', ax=axes);
df.plot(kind='line', x='year', y='men_advanced_degree', ax=axes);
```

```python
axes = plt.gca()
plt.title("Women Hourly Wage Based on Education from 1973 to 2022") 
df.plot(kind='line', x='year', y='women_less_than_hs', ax=axes);
df.plot(kind='line', x='year', y='women_high_school', ax=axes);
df.plot(kind='line', x='year', y='women_some_college', ax=axes);
df.plot(kind='line', x='year', y='women_bachelors_degree', ax=axes);
df.plot(kind='line', x='year', y='women_advanced_degree', ax=axes);
```

```python
'/ US Hourly Wage by gender/'

selected_years = [1982, 1992, 2002, 2012, 2022]
df_selected_years = df[df['year'].isin(selected_years)]
df_selected_years[['year', 'men_high_school', 'women_high_school']].sort_values(by='year').plot(x='year', kind='bar')
plt.title('Hourly Wage after a High School Degree Based on Gender')
plt.xlabel('Year')
plt.ylabel('Hourly Wage ($)')
plt.show()
```

```python
selected_years = [1982, 1992, 2002, 2012, 2022]
df_selected_years = df[df['year'].isin(selected_years)]
df_selected_years[['year', 'men_bachelors_degree', 'women_bachelors_degree']].sort_values(by='year').plot(x='year', kind='bar')
plt.title('Hourly Wage after a Bachelor Degree Based on Gender')
plt.xlabel('Year')
plt.ylabel('Hourly Wage ($)')
plt.show()
```

```python
selected_years = [1982, 1992, 2002, 2012, 2022]
df_selected_years = df[df['year'].isin(selected_years)]
df_selected_years[['year', 'men_advanced_degree', 'women_advanced_degree']].sort_values(by='year').plot(x='year', kind='bar')
plt.title('Hourly Wage after an Advanced Degree Based on Gender')
plt.xlabel('Year')
plt.ylabel('Hourly Wage ($)')
plt.show()
```

```python
'/ US Hourly Wage by race/'

selected_years = [1982, 1992, 2002, 2012, 2022]
df_selected_years = df[df['year'].isin(selected_years)]
df_selected_years[['year', 'black_high_school', 'hispanic_high_school', 'white_high_school']].sort_values(by='year').plot(x='year', kind='bar')
plt.title('Hourly Wage after a High School Degree Based on Race')
plt.xlabel('Year')
plt.ylabel('Hourly Wage ($)')
plt.show()
```
